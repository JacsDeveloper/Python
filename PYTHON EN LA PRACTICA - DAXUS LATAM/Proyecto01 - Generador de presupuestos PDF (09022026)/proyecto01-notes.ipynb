# Creando y generando PDF

## Pasos para crear el PDF: 

* Crear el PDF
* Definir el tipo de letra
* Llenamos el documento
* Guardar

```python
# Instalando la biblioteca fpdf para crear un archivo PDF
# !pip install fpdf

!pip install fpdf
```

```python
# Llamando la librería
from fpdf import FPDF

# Guardamos los datos ingresados por el usuario en variables para usarlos posteriormente
nombre_proyecto = input("Ingrese nombre del proyecto: ")
horas_estimadas = input("Ingrese horas estimadas para el proyecto: ")
valor_hora = input("Ingrese valor por hora trabajada: ")
plazo_estimado = input("Ingrese plazo estimado para el proyecto (en días): ")

calculo_pago = int(horas_estimadas) * int(valor_hora)

# Asignando la clase FPDF a una variable
pdf = FPDF()
# Agregando una página al PDF
pdf.add_page()
# Definiendo el tipo de letra y tamaño
pdf.set_font("Arial", size=12)

# Cargando la imagen de fondo para el PDF (Eje X=0, Eje Y=0 para que quede de fondo)
pdf.image("Template.png", x=0, y=0)

# Agregando el texto al PDF (Eje X=115, Eje Y=145 para que quede en el espacio del template)
pdf.text(115, 145, nombre_proyecto)
pdf.text(115, 160, horas_estimadas)
pdf.text(115, 175, valor_hora)
pdf.text(115, 190, plazo_estimado)
pdf.text(115, 205, str(calculo_pago))

# Generando el PDF 
pdf.output("proyecto01-presupuesto.pdf")
print("¡Archivo PDF generado exitosamente!")
```
