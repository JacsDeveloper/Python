# Imprimimos el valor máximo y mínimo de cierre de la Acción de Apple según la data

```python
# Generamos toda la secuencia del código anterior hasta acá.
import yfinance
import matplotlib

# Obtenemos los datos de las acciones de Apple (AAPL) desde Yahoo Finance (de 6 meses atrás hasta hoy)
data = yfinance.Ticker("AAPL").history(period="6mo")
cierre = data.Close

# Cotización máxima de las acciones de Apple (AAPL) en los últimos 6 meses
max_cierre = cierre.max()
min_cierre = cierre.min()
valor_medio = cierre.mean()
print("El valor máximo es:", max_cierre)
print("El valor mínimo es:", min_cierre)
print("El valor promedio es:", valor_medio)
print("-----------------------------------------------")

# Round es para redondear a 2 decimales
print("El valor máximo es:",round(max_cierre, 2))
print("El valor mínimo es:",round(min_cierre, 2))
print("El valor promedio es:",round(valor_medio, 2))
print("-----------------------------------------------")

# También podríamos redondear desde que asignamos el valor a la variable
max_cierre = round(cierre.max(), 2)
min_cierre = round(cierre.min(), 2)
valor_medio = round(cierre.mean(), 2)
print("El valor máximo es:", max_cierre)
print("El valor mínimo es:", min_cierre)
print("El valor promedio es:", valor_medio)
```

# Enviar correo electrónico automáticamente

Aquí se definen los pasos que hacemos manualmente para enviar un correo paso a paso pero lo vamos a hacer con Python.

* Abrir una nueva pestaña del navegador
* Entrar al correo
* Clic en Redactar
* Escribimos dirección del destinatario
* Escribimos asunto
* Redactamos el correo (Cuerpo del correo)
* Clic en Enviar

```python
# Importamos las librerías
import pyautogui
import pyperclip
# Esta librería viene por defecto y es para manejar las opciones del navegadror
import webbrowser

# Abrimos el navegador en la página de Gmail
webbrowser.open("https://mail.google.com/mail/u/0/#inbox")
```

## La metodología que aquí se aplicó es determinando la posición del mouse sobre el botón de Redactar para llamar la función Click() e hiciera la acción de dar clic en una posición específica

```python
# La librería pyautogui es para manipular el HardWare del PC (automatizar el teclado y el mouse), 
# y la librería pyperclip es para copiar y pegar texto utilizando el portapapeles del sistema operativo.

# Pedimos la posición del mouse para saber dónde hacer clic
# Para este ejemplo lo posicionó en el botón "Redactar" de Gmail, pero cada quien lo puede posicionar donde quiera
pyautogui.position()
```

```python
# Con la ventana de Chrome maximizada, el botón "Redactar" de Gmail se encuentra en la posición (115, 274)
# Entonces, con el método click de pyautogui hacemos clic en esa posición para abrir la ventana de redacción de correo.

# Retomamos el código anterior y lo continuamos.
# Importamos las librerías
import pyautogui
import pyperclip
# Esta librería viene por defecto y es para manejar las opciones del navegadror
import webbrowser

# Definimos el destinatario, asunto y mensaje que queremos enviar por correo
destinatario = "destinatario@ejemplo.com"
asunto = "Asunto del correo"
mensaje = """Cordial saludo. 

Este es un correo de prueba enviado desde Python utilizando las librerías pyautogui y pyperclip.

Los datos de las acciones de Apple (AAPL) en los últimos 6 meses son los siguientes:
Valor máximo: {max_cierre}
Valor mínimo: {min_cierre}
Valor promedio: {promedio_cierre}

Saludos,
Jacs
"""

# Abrimos el navegador en la página de Gmail
webbrowser.open("https://mail.google.com/mail/u/0/#inbox")

# Como las líneas las rrecorre en microsegundos, hacemos una pausa de 3 segundos para que el navegador tenga tiempo de abrirse antes del clic.
pyautogui.sleep(3)
pyautogui.PAUSE = 1.5

# Hacemos clic en el botón "Redactar" de Gmail para abrir la ventana de redacción de correo
pyautogui.click(115, 274)

# Copiamos el destinatario al portapapeles y lo pegamos en el campo "Para" del correo, luego damos Tab para pasar al campo "Asunto".
pyperclip.copy(destinatario)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# Asunto del correo
pyperclip.copy(asunto)
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# Cuerpo del correo
pyperclip.copy(mensaje)
pyautogui.hotkey("ctrl", "v")

# Enviar correo
pyautogui.hotkey("ctrl", "Enter")

```
