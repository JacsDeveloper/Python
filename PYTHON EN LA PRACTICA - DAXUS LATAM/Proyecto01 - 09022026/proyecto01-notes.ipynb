# Guardando en variables

```python
nombre_proyecto = input("Ingrese nombre del proyecto: ")
horas_estimadas = input("Ingrese horas estimadas para el proyecto: ")
valor_hora = input("Ingrese valor por hora trabajada: ")
plazo_estimado = input("Ingrese plazo estimado para el proyecto (en días): ")
```

```python
print("El nombre del proyecto es:", nombre_proyecto)
print("Las horas estimadas para el proyecto son:", horas_estimadas)
print("El valor por hora trabajada es:", valor_hora)
print("El plazo estimado para el proyecto es:", plazo_estimado, "días")
```

## Cálculos con Python

```python
print("El valor a pagar es:", int(horas_estimadas) * int(valor_hora))
```
